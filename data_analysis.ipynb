# Data Analysis

```python
import pandas as pd
import plotly.graph_objs as go

pd.options.plotting.backend = "plotly"
```

### Weather

First we analyse the file `Données météo Ferrières (horaire).xlsx`

```python
path_folder = 'Data/Weather/'
df_weather_horaire = pd.read_excel(path_folder +'Données météo Ferrières (horaire).xlsx')
```

```python
print(f'Data from {df_weather_horaire.date.min()} to {df_weather_horaire.date.max()}')
df_weather_horaire.describe().style.format("{:.2f}")
```

```python
fig = df_weather_horaire.plot(x='date', y=df_weather_horaire.columns[-3:], title='Temperature')
fig.show()
```

```python
fig = df_weather_horaire.plot(x='date', y=df_weather_horaire.columns[1:5], title='Temperature')
fig.show()
```

```python
path_folder = 'Data/Weather/'
df_weather_daily = pd.read_excel(path_folder +'Données météo Ferrières (journalière)).xlsx', header=[0,1])
df_weather_daily.describe().style.format("{:.2f}")
```

```python
df_weather_daily
```

```python
df_weather_daily.columns
```

```python
fig = go.Figure()
# fig.add_trace(go.Scatter(x=df_weather_daily[('Unnamed: 0_level_0', 'Date/heure')], y=df_weather_daily[('Précipitations [mm]','Somme')], name='Précipitations [mm]'))
for i in range(1,7):    
    fig.add_trace(go.Scatter(x=df_weather_daily[('Unnamed: 0_level_0', 'Date/heure')], y=df_weather_daily[(f'EAG Humidité du sol {i} [%]', 'moy')]/100, name=f'Humidity of soil {i}'))
    fig.update_layout(title=f'Temperature of soil {i}', xaxis_title='Date', yaxis_title='[%]')
fig.update_yaxes(tickformat="%{n}f")
fig.show()
```

```python
for i in range(1,7):
    fig = go.Figure()
    # fig.add_trace(go.Scatter(x=df_weather_daily[('Unnamed: 0_level_0', 'Date/heure')], y=df_weather_daily[('Précipitations [mm]','Somme')], name='Précipitations [mm]'))
    cols = [col for col in df_weather_daily.columns if f'Température du sol {i}' in col[0]]
    for col in cols:
        fig.add_trace(go.Scatter(x=df_weather_daily[('Unnamed: 0_level_0', 'Date/heure')], y=df_weather_daily[col], name=f'T° {col[1]} of soil {i}'))
    fig.update_layout(title=f'Temperature of soil {i}', xaxis_title='Date', yaxis_title='[°C]')
    fig.show()
```
