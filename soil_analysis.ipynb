#### Irrigation analysis

Irrigation using different types of soil

```python
irri = IrrigationAnalysis(dates = ('2022-04-01', '2022-09-15'), crop_type='Potato', soil_class = 'SIL')
```

```python
soil_names = ['S', 'LS', 'SL', 'L', 'SIL', 'SI', 'SCL', 'CL', 'SICL', 'SC', 'SIC', 'C']
soil_irrigation = {}
for soil in soil_names:
    irri.soil = soil
    result_df = irri.results(type_ir='A')
    # soil_irrigation['Date'] = list(result_df['Date'])
    soil_irrigation[soil] = result_df.values

soil_irrigation['Date'] = result_df.keys()

soil_irrigation = pd.DataFrame(soil_irrigation)
soil_irrigation
```

```python
# Plot the multiline plot
plt.figure(figsize=(10,8))
palette = sns.color_palette("husl", n_colors=soil_irrigation.shape[1]-1)
sns.lineplot(x="Date", y="value", hue="variable", data=pd.melt(soil_irrigation, ['Date']), palette=palette)

# Show the plot
plt.title("Irrigation for different types of soil")
plt.xlabel("Date")
plt.ylabel("Irrigation (mm)")
plt.legend(title="Soil type")
plt.show()

#Barplot
soils = soil_irrigation.loc[:, soil_irrigation.columns != 'Date']
sum_irri = soils.sum()[soils.sum().argsort()]
palette = sns.color_palette("husl", n_colors=soil_irrigation.shape[1]-1)
palette = [palette[i] for i in soils.sum().argsort()]
plt.figure(figsize=(10,8))
sns.barplot(x=sum_irri.keys(),y=sum_irri.values, palette=palette)

# Show the plot
plt.title("Irrigation for different types of soil")
plt.xlabel("Soil")
plt.ylabel("Irrigation sum (mm)")
plt.show()
```

```python
# Plot the multiline plot
plt.figure(figsize=(10,8))
sns.lineplot(x="Date", y="value", hue="variable", data=pd.melt(soil_irrigation[['Date', 'SICL', 'SCL']], ['Date']))

# Show the plot
plt.title("Maximun and minimun indication")
plt.xlabel("Date")
plt.ylabel("Irrigation (mm)")
plt.legend(title="Soil type")
plt.show()
```
